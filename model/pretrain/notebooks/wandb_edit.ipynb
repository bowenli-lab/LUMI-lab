```python
import wandb
import re
api = wandb.Api()
runs = api.runs("sdl-lnp-team/1920-model-kfold-nested-benchmark-sweep",)

for run in runs:
    pattern = (r'(?P<model_type>^[a-zA-Z0-9-]+)_'               # Model type
           r'(?P<size>[a-zA-Z0-9-]+)_epoch-'                # Size
           r'lr(?P<learning_rate>[0-9e-]+)-cw'              # Learning rate
           r'(?P<cw>[0-9.]+)-lg'                            # CW
           r'(?P<lg>[0-9.]+)_'                              # LG
           r'(?P<date>\d{4}-\d{2}-\d{2})_fold_'             # Date
           r'(?P<fold>\d+)$')                               # Fold
    match = re.match(pattern, run.name)
    # print(match.groupdict())
    run.config.update(match.groupdict())
    run.update() # This line sends the updated information over to your W&B Dashboard
    print(run.name, run.state)
    # break
```
