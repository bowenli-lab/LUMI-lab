```python
import pandas as pd
```

```python
data_path = '/h/pangkuan/dev/SDL-LNP/model/data_process/220k_library.csv'

df = pd.read_csv(data_path)

df
```

```python
# split to NUM files
SAVE_ADDRESS = "/scratch/ssd004/datasets/cellxgene/3d_molecule_data/220k-lib/partitioned_csv/"
length = len(df)
NUM =10
step = length // NUM
for i in range(NUM):
    
    print("processing ", i)
    if i == NUM - 1:
        df.iloc[i*step:].to_csv(SAVE_ADDRESS + f'{i}.csv', index=False)
    else: 
        df.iloc[i*step:(i+1)*step].to_csv(SAVE_ADDRESS + f'{i}.csv', index=False)
```
