```python
from chembl_webresource_client.new_client import new_client
import pandas as pd
# Define the client
molecule = new_client.molecule

# Query for molecules with MW between 500 and 1500
# molecules = molecule.filter(molecule_properties__mw_freebase__gt=500,
#                             molecule_properties__mw_freebase__lt=1500)

# get all molecules
molecules = molecule.filter()
```

```python
def process_pka(most_apka, most_bpka):
    if most_apka is None and most_bpka is None:
        return None
    if most_apka is None:
        return (most_bpka, most_bpka)
    if most_bpka is None:
        return (most_apka, most_apka)
    return (most_apka, most_bpka)
    
```

```python
df = pd.DataFrame(molecules)
df
```

```python
res = []
for mol in molecules:
    if 'molecule_structures' not in mol or not mol['molecule_structures']:
        continue
    smiles = mol['molecule_structures']['canonical_smiles']
    most_apka = mol['molecule_properties']['cx_most_apka']
    most_bpka = mol['molecule_properties']['cx_most_bpka']
    processed_pka = process_pka(most_apka, most_bpka)
    if processed_pka is not None:
        res.append({
            "smiles": smiles,
            "most_apka": processed_pka[0],
            "most_bpka": processed_pka[1]
        })
```
