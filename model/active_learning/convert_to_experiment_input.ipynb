````python
# read in selected candidate csv
SELECTED_CANDIDATE_PATH = "selected_structures_exp9.csv"
selected_candidate_df = pd.read_csv(SELECTED_CANDIDATE_PATH)
selected_candidate_df
````

````python
# create new column for reagent id, to match the Reagent column in the mapping table
reagent_ids = {
    "A": [],
    "B": [],
    "C": [],
    "D": []
}
for i in range(len(selected_candidate_df)):
    for reagent in ["A", "B", "C", "D"]:
        name = selected_candidate_df[f"{reagent}_smiles"].iloc[i]
        reagent_id = reagent_mapping_df[reagent_mapping_df["Reagent SMILES"] == name]["Reagent"].values
        assert len(reagent_id) == 1
        reagent_ids[reagent].append(reagent_id[0])
for reagent in ["A", "B", "C", "D"]:
    selected_candidate_df[f"{reagent}_id"] = reagent_ids[reagent]
selected_candidate_df
````

````python
# put the 184 lipids in selected_candidate_df into two plates
assert len(selected_candidate_df) <= 184
first_plate = selected_candidate_df.iloc[:92]
second_plate = selected_candidate_df.iloc[92:]

# generate the experiment input json
experiment_input = []
for plate in [first_plate, second_plate]:
    targets = []
    for i in range(len(plate)):
        target = {
            "lipid": {
                "amines": plate["A_id"].iloc[i],
                "isocyanide": plate["B_id"].iloc[i],
                "lipid_carboxylic_acid": plate["D_id"].iloc[i],
                "lipid_aldehyde": plate["C_id"].iloc[i]
            }
        }
        targets.append(target)
    experiment_input.append({"targets": targets})

experiment_input[0]
````

````python
# save to json
import json

with open("experiment_input.json", "w") as f:
    json.dump(experiment_input, f, indent=4)
````
